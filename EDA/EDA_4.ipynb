도착항구와 선박 국가 같은 경우

```python
import warnings
warnings.filterwarnings(action="ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_parquet("../../data/HD_data/train.parquet").drop(columns="SAMPLE_ID")
```

```python
train_df["ARI_CO"].unique()
```

```python
sorted(train_df["FLAG"].unique())
```

```python
num_rows = 1
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 5))

df_1 = train_df[(train_df["ARI_CO"] == "CN") | ("China" == train_df["FLAG"])]
ax1 = sns.histplot(df_1["CI_HOUR"], bins=50, ax=axes[0])
ax1.set_title("same country")

df_2 = train_df[(train_df["ARI_CO"] != "CN") | ("China" == train_df["FLAG"])]
ax2 = sns.histplot(df_2["CI_HOUR"], bins=50, ax=axes[1])
ax2.set_title("not same country")

fig.suptitle("CN")
plt.show()
```

```python
num_rows = 1
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 5))

df_1 = train_df[(train_df["ARI_CO"] == "KR") | ("Korea" in train_df["FLAG"])]
ax1 = sns.histplot(df_1["CI_HOUR"], bins=50, ax=axes[0])
ax1.set_title("same country")

df_2 = train_df[(train_df["ARI_CO"] != "KR") | ("Korea" in train_df["FLAG"])]
ax2 = sns.histplot(df_2["CI_HOUR"], bins=50, ax=axes[1])
ax2.set_title("not same country")

fig.suptitle("KR")
plt.show()
```

선박 국가와 도착 항구가 같던 다르던 크게 상관이 없는 듯 하다

```python
num_rows = 1
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 5))

df_1 = train_df[(train_df["ARI_CO"] == "SG") | ("Singapore" == train_df["FLAG"])]

df_2 = train_df[(train_df["ARI_CO"] != "SG") | ("Singapore" == train_df["FLAG"])]

# 겹치는 히스토그램 그래프 추가
ax1 = sns.histplot(df_1["CI_HOUR"], bins=50, ax=axes[0], color='red', alpha=0.5, label="same country")
ax2 = sns.histplot(df_2["CI_HOUR"], bins=50, ax=axes[0], color='blue', alpha=0.5, label="not the same country")
ax1.legend()
ax2.legend()
# 분포 그래프 추가

sns.kdeplot(df_1["CI_HOUR"], ax=axes[1], color='red', linestyle='--')
sns.kdeplot(df_2["CI_HOUR"], ax=axes[1], color='blue', linestyle='--')


fig.suptitle("Singapore")
plt.show()
```

```python
num_rows = 1
num_cols = 2

fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 5))

df_1 = train_df[(train_df["ARI_CO"] == "VE") | ("China" == train_df["FLAG"])]

df_2 = train_df[(train_df["ARI_CO"] != "VE") | ("China" == train_df["FLAG"])]

# 겹치는 히스토그램 그래프 추가
sns.histplot(df_1["CI_HOUR"], bins=50, ax=axes[0], color='red', alpha=0.5, label="same country")
sns.histplot(df_2["CI_HOUR"], bins=50, ax=axes[0], color='blue', alpha=0.5, label="not the same country")

# 분포 그래프 추가

sns.kdeplot(df_1["CI_HOUR"], ax=axes[1], color='red', linestyle='--')
sns.kdeplot(df_2["CI_HOUR"], ax=axes[1], color='blue', linestyle='--')

fig.suptitle("China")
plt.show()
```
