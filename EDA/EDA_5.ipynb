날짜 시간에 따른 CI HOUR분포

```python
import warnings
warnings.filterwarnings(action="ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_parquet("../data/train.parquet").drop(columns="SAMPLE_ID")
```

```python
train_df['ATA'] = pd.to_datetime(train_df['ATA'])

# datetime을 여러 파생 변수로 변환
train_df['year'] = train_df['ATA'].dt.year
train_df['month'] = train_df['ATA'].dt.month
train_df['day'] = train_df['ATA'].dt.day
train_df['hour'] = train_df['ATA'].dt.hour
train_df['minute'] = train_df['ATA'].dt.minute
train_df['weekday'] = train_df['ATA'].dt.weekday
```

```python
df = train_df[["year", "CI_HOUR"]]
sns.displot(df, x="year", y="CI_HOUR")
plt.show()
```

```python
df = train_df[["month", "CI_HOUR"]]
sns.displot(df, x="month", y="CI_HOUR")
plt.show()
```

```python
df = train_df[["day", "CI_HOUR"]]
sns.displot(df, x="day", y="CI_HOUR")
plt.show()
```

```python
df = train_df[["hour", "CI_HOUR"]]
sns.displot(df, x="hour", y="CI_HOUR")
plt.show()
```

```python
df = train_df[["ATA_LT", "CI_HOUR"]]
sns.displot(df, x="ATA_LT", y="CI_HOUR")
plt.show()
```
